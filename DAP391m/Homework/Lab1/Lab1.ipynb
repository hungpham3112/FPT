## Question 10

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
numerical_df = df._get_numeric_data()
correlation_matrix = numerical_df.corr()
condition = correlation_matrix['price'].abs() > 0.6
high_correlation_cols = correlation_matrix[condition].index
high_correlation_df = df[high_correlation_cols]
```

```python
x_train, x_test, y_train, y_test = train_test_split(high_correlation_df.drop(['price'], axis=1), high_correlation_df['price'], test_size=0.3, random_state=312)
```

```python
ridge = Ridge(alpha=0.1)
ridge.fit(x_train, y_train)
ridge.score(x_test, y_test)
```

## Question 11

```python
lst1 = [("scale", StandardScaler()), ("polynomial", PolynomialFeatures(degree=2, include_bias=False)), ("model", Ridge(alpha=0.1))]
pipe1 = Pipeline(lst1)
```

```python
pipe1.fit(x_train, y_train)
```

```python
pipe1.score(x_test, y_test)
```
