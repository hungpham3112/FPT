```python
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_image(name: str, image: np.ndarray):
    cv2.imshow(name, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def extract_channel(image: np.ndarray, channel: int):
    """
    0: Blue
    1: Green
    2: Red
    """
    array = np.zeros_like(image)
    array[:, :, channel] = cv2.split(image)[channel]
    return array
```

```python
image = cv2.imread("./lenna.png")
type(image)
```

```python
read_image("Image", image)
```

```python
read_image("Blue", extract_channel(image, 0))
```

```python
red = cv2.calcHist([image], [2], None, [256], [0, 256])
green = cv2.calcHist([image], [1], None, [256], [0, 256])
blue = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.plot(red, color='red', label='Red')
plt.plot(green, color='green', label='Green')
plt.plot(blue, color='blue', label='Blue')
plt.title('Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```
