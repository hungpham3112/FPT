```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
img = Image.open('lenna.png')
plt.imshow(img)
plt.show()
```

```python
img_arr = np.array(img)
print(img_arr.shape)
```

```python
R = img_arr[:,:,0].max()
G = img_arr[:,:,1].max()
B = img_arr[:,:,2].max()
```

```python
print(R,G,B)
```

```python
cb = np.array([[225/R,0,0],
              [0,255/G,0],
              [0,0,255/B]])
```

```python
img_arr_1 = img_arr.copy()
for row in range(img_arr_1.shape[0]):
    for col in range (img_arr_1.shape[1]):
        img_arr_1[row,col] = np.dot(cb,img_arr[row,col].T)
print( img_arr_1[0,0])    
```

```python
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(img_arr_1)
plt.show()
```
